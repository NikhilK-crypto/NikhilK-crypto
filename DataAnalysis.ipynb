## Data Visualization 

### bargraph of top 10 countries with high count of in the dataset

```python
plt.figure(figsize=(6,5), dpi=100)
df.Country.value_counts()[:10].plot(kind='bar', color='#32a893')

plt.title('Count of countries', fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('Country Name', fontdict={'fontname':'Comic Sans MS','fontsize':16})
plt.ylabel('Count', fontdict={'fontname':'Comic Sans MS','fontsize':16})

plt.legend()
plt.show()
```

### income distribution using histogram

```python
plt.figure(figsize=(8,5),dpi=100)

plt.hist(df.Income,color='#a89632')

plt.title('Income Distribution')
plt.xlabel('Income in USD')
plt.ylabel('No of People')

plt.yticks(range(0,125,10))
plt.xticks(range(1000,65000,5000))

plt.show()
```

### Income Comparision between male and female employees

```python
df=df.dropna()
```

```python
df.isnull().values.any()
```

```python
male_income=df.loc[df.gender=='Male']['Income']
female_income=df.loc[df.gender=='Female']['Income']
```

```python
print(female_income.count())
male_income.count()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(4,3),dpi=200)

plt.boxplot([male_income,female_income], labels=['Male Income','Female Income'])

plt.title('Income Comparision')
plt.ylabel('Anual Income in USD')

plt.yticks(range(0,70000,5000))

plt.show()
```

### 10 countries with high and low earning employees

```python
a=df.groupby(['Country'],as_index=False).Income.mean().sort_values(by='Income', ascending = False).head(10)
b=df.groupby(['Country'],as_index=False).Income.mean().sort_values(by='Income', ascending = False).tail(10)
```

```python
a=a.set_index('Country')
b=b.set_index('Country')
```

```python
print(a)
print('\n')
print(b)
```

```python
plt.figure(figsize=(8,5),dpi=100)
a.plot(kind='barh', color='#3498eb')
plt.title('Top 10 high earning Countries',fontdict={'fontsize':'20'})
plt.xlabel('Average income in USD', fontdict={'fontsize':'16'})
plt.ylabel('Countries', fontdict={'fontsize':'16'})
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.figure(figsize=(4,3),dpi=100)

b.plot(kind='barh', color='#9994d1')
plt.title('Top 10 Low Earning Countries', fontdict={'fontsize':'20'})
plt.xlabel('Average income in USD', fontdict={'fontsize':'16'})
plt.ylabel('Countries', fontdict={'fontsize':'16'})
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

### 10 jobs with high and low income 

```python
x=df.groupby(['Job'],as_index=False).Income.mean().sort_values(by='Income', ascending = False).head(10)
y=df.groupby(['Job'],as_index=False).Income.mean().sort_values(by='Income', ascending = False).tail(10)
```

```python
x=x.set_index('Job')
y=y.set_index('Job')
```

```python
print(x)
print('\n')
print(y)
```

```python
x.plot(kind='barh', color='#a4c7a3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.title('Top 10 Jobs with high earning',fontdict={'fontsize':'20'})
plt.xlabel('Average income in USD', fontdict={'fontsize':'16'})
plt.ylabel('Job Title', fontdict={'fontsize':'16'})

plt.show()
```

```python
plt.figure(figsize=(8,5),dpi=100)

y.plot(kind='barh', color='#c7c6a3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.title('Top 10 Jobs with low earning',fontdict={'fontsize':'20'})
plt.xlabel('Average income in USD', fontdict={'fontsize':'16'})
plt.ylabel('Job Title', fontdict={'fontsize':'16'})

plt.show()
```
